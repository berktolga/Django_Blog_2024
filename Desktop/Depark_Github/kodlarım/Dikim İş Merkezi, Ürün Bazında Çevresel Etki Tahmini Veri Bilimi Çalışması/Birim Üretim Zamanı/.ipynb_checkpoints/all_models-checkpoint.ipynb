```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_percentage_error
import xgboost as xgb
import catboost as cb
import lightgbm as lgb
import matplotlib.pyplot as plt

# Veriyi ykleyelim
data_path = 'birim_uretim_zamani_duzenlenmis.csv'
data = pd.read_csv(data_path)

# zellik ve hedef deikenleri belirleme
# Uretim, Etut, BirimUretimZamani dndaki kolonlar kategorik kabul ediliyor
X = data[['ModelAdi', 'PersonelNo', 'ModelTipAdi', 'OperAdi', 'MakinaTipi', 'Uretim','SIPARIS_ADEDI','GIYSI_GRUBU','GIYSI_CINSI','GIYSI_CINSI_DETAYI']]
#Versiyon4'te OrderNo,PersonelNo,ModelTipAdi,OperAdi,MakinaTipi,Uretim,UretimTarihi
y = data['BirimUretimZamani']  # Hedef deiken

# Kategorik deikenleri saysallatrma (One-hot encoding)
# Uretim ve Etut dndaki tm kolonlar iin one-hot encoding uyguluyoruz
X = pd.get_dummies(X, columns=['ModelAdi', 'PersonelNo', 'ModelTipAdi', 'OperAdi', 'MakinaTipi','GIYSI_GRUBU','GIYSI_CINSI','GIYSI_CINSI_DETAYI'])

# Veriyi leklendirme (Sadece saysal zellikleri leklendirelim)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Modelleri tanmlayalm
models = {
    'Random Forest': RandomForestRegressor(random_state=42, n_jobs=-1),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42),
    'Elastic Net': ElasticNet(random_state=42),
    'SVR': SVR(),
    'Neural Network': MLPRegressor(random_state=42, max_iter=1000),
    'KNN': KNeighborsRegressor(n_jobs=-1),
    'xGBoost': xgb.XGBRegressor(random_state=42, n_jobs=-1),
    'CatBoost': cb.CatBoostRegressor(random_state=42, silent=True),
    'LightGBM': lgb.LGBMRegressor(random_state=42, n_jobs=-1)
}

# Her model iin sonular tutmak iin bo bir DataFrame
results = pd.DataFrame(columns=['Model', 'MAPE'])

# Tm modelleri eitip deerlendiriyoruz
for model_name, model in models.items():
    # Modeli eit
    model.fit(X_scaled, y)
    
    # Tahmin yap
    y_pred = model.predict(X_scaled)
    
    # MAPE hesaplayalm
    mape = mean_absolute_percentage_error(y, y_pred)
    
    # Sonular bir DataFrame olarak oluturup sonular DataFrame'ine ekleyelim
    result_row = pd.DataFrame({
        'Model': [model_name],
        'MAPE': [mape]
    })
    results = pd.concat([results, result_row], ignore_index=True)

# Sonularn tablo olarak gsterilmesi
print(results)

# Tm veriyi kullanarak modelleri eitip tahminleri yapalm ve grafikleri izelim
for model_name, model in models.items():
    # Modeli eit
    model.fit(X_scaled, y)
    
    # Tahmin yap
    y_pred = model.predict(X_scaled)
    
    # Gerek ve tahmin edilen deerleri iz
    plt.figure(figsize=(10, 6))
    plt.plot(y.values, label='Actual', color='blue', alpha=0.6)
    plt.plot(y_pred, label='Forecast', color='red', alpha=0.6)
    
    # Grafiin balk ve etiketleri
    plt.title(f'{model_name} - BirimUretimZamani')
    plt.xlabel('Sample')
    plt.ylabel('BirimUretimZamani')
    plt.legend()
    plt.show()
```
